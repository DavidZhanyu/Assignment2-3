```python
#Show a breakdown of distance from home by job role and attrition.

sns.scatterplot(data=df, x='DistanceFromHome', y='JobRole', hue='Attrition')
```

```python
sns.countplot(data=df, x='Attrition')
```

```python
sns.countplot(data=df, x='JobRole')
```

```python
sns.histplot(data=df, x='DistanceFromHome')
```

```python
# Create a pivot table to show the average distance from home by job role and attrition
pivot_table = df.pivot_table(values='DistanceFromHome', index='JobRole', columns='Attrition', aggfunc='mean')

print(pivot_table)
```

```python
# Create a pivot table to show the average monthly income by education and attrition
pivot_table = df.pivot_table(values='MonthlyIncome', index='Education', columns='Attrition', aggfunc='mean')

print(pivot_table)
```

```python
sns.barplot(x='DistanceFromHome', y='JobRole', hue='Attrition', data=df)
```

```python
sns.barplot(x='MonthlyIncome', y='EducationField', hue='Attrition', data=df)
plt.show()
```
